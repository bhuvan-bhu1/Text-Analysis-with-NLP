```python
pos_neg = pd.read_excel('/content/drive/MyDrive/Sist/Positive and Negative Word List.xlsx')
pos_neg.columns
```

```python
negetive_words = [i for i in pos_neg['Negative Sense Word List'].to_list() if i != 'nan']
positive_words = [i for i in pos_neg['Positive Sense Word List'].to_list() if i != 'nan']
```

```python
len(positive_words)
```

```python
len(negetive_words)
```

```python
positive_score = []
negetive_score = []
polarity_score = []
subjectivity_score = []
for data in df['data_list']:
  tokens = []
  req_list = []
  for row in data:
        tokens += word_tokenize(row) #creating tokens
  for word in tokens:
    if word not in sw_nltk:  #Stopwords removing
      req_list.append(ps.stem(word))

  pos = 0
  neg = 0
  for word in req_list:
      if word in positive_words:
          pos += 1
      elif word in negetive_words:
          neg -= 1
  neg *= -1
  #Calculating Polarity and Subjectivity scores
  polarity = (pos - neg) / ((pos + neg) + 0.000001)
  subjectivity = (pos + neg) / (len(req_list) + 0.000001)
  positive_score.append(round(pos,4))
  negetive_score.append(round(neg,4))
  polarity_score.append(round(polarity,4))
  subjectivity_score.append(round(subjectivity,4))
```

```python
print(len(positive_score))
print(len(negetive_score))
print(len(polarity_score))
print(len(subjectivity_score))
```

```python
#Adding the calculated values into the Dataframe
df['POSITIVE SCORE'] = positive_score
df['NEGATIVE SCORE'] = negetive_score
df['POLARITY SCORE'] = polarity_score
df['SUBJECTIVITY SCORE'] = subjectivity_score
```

```python
#Calculating Average Sentence Length
average_sentence_length = []
for i in range(len(df)):
  words = 0
  sentences = 0
  for row in df.iloc[i]['text']:
    words += len([i for i in word_tokenize(row) if i not in punctuation_list])
    sentences += len(sent_tokenize(row))
  try:
    average_sentence_length.append(round(words / sentences,4))
  except:
    average_sentence_length.append(0)
```

```python
#Declaring syllable dictionary
syllable_dict = cmudict.dict()
len(syllable_dict)
```

```python
#Function to count number syallables in a given word
def count_syllables(word):
  if word.lower() not in syllable_dict:
      return 0
  return [len(list(y for y in x if y[-1].isdigit())) for x in syllable_dict[word.lower()]][0]
```

```python
#Calculating complex words percentage in the corpus
percentage_complex_words = []
for i in range(len(df)):
  req_list = []
  for row in df.iloc[i]['data_list']:
    req_list += word_tokenize(row)
  count = 0
  for word in req_list:
    if count_syllables(word) and word not in sw_nltk:
      count += 1
  try:
    percentage_complex_words.append(round(count / len(req_list),4))
  except:
    percentage_complex_words.append(0)
```

```python
#Calculating Fog index of corpus
def calculate_fog_index(average_sentence_length, percentage_complex_words):
    fog_index = 0.4 * (average_sentence_length + percentage_complex_words)
    return round(fog_index,4)

df['FOG INDEX'] = np.nan
```

# Syllable Count Per Word

```python
#Function to calculate number of syllables in a word
def count_syllables(word):
  word = word.lower()
  num_vowels = len([char for char in word if char in 'aeiou'])
  if word.endswith('es') or word.endswith('ed'):
    num_vowels -= 1
  return num_vowels
```

```python
#Calculating syllable of every word in the corpus
word_list = []
syllable_count_per_word = []
for i in range(len(df)):
  words = []
  for row in df.iloc[i]['data_list']:
    words += [i for i in word_tokenize(row) if i not in punctuation_list]
  word_list.append(words)
  syllable_count_per_word.append([count_syllables(word) for word in words])
```

```python
#Adding the column to the Dataframe
df['SYLLABLE PER WORD'] = syllable_count_per_word
df.head(3)
```

# Personal Pronouns

```python
#Declaring the pattern to be consider for calculating the count of personal pronouns
pattern = r"\b(I|we|my|ours|us)\b"
pattern = r"(?<!\bUS\b)" + pattern # US vs us
pattern
```

```python
#Calculating personal pronouns of the words in each sentences
pronouns_list = []
for i in range(len(df)):
  count = 0
  for row in df.iloc[i]['data_list']:
    count += len(re.findall(pattern, row, flags=re.IGNORECASE))
  pronouns_list.append(count)
len(pronouns_list)
```

```python
#Adding the column to the Dataframe
df['PERSONAL PRONOUNS'] = pronouns_list
df.head(3)
```

```python
#Calculating the Average Word Length of the Corpus
average_word_length = []
for i in range(len(df)):
  words = []
  for row in df.iloc[i]['data_list']:
    words += [word for word in word_tokenize(row)]
  total_char = sum([len(word) for word in words])
  try:
    average_word_length.append(round(total_char / len(words),4))
  except:
    average_word_length.append(0)
len(average_word_length)
```
